```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('heart.csv')
```

```python
data
```

# MAPA DE CALOR: REPRESENTA A CORRELAÇÃO QUE CADA FEATURE TEM COM AS OUTRAS

```python
plt.figure(figsize=(12,8))
sns.heatmap(data.corr(),annot=True)
```

# GRÁFICO DE DISTRIBUIÇÃO NORMAL

```python
for feature in data:
    plt.figure(figsize=(12,8))
    sns.histplot(data[feature], stat='density', color='blue')
    sns.kdeplot(data[feature],color='r')
    plt.axvline(data[feature].mean(),linestyle='--',linewidth=0.9,color='y')
    min_ylim,max_ylim =plt.ylim()
    plt.text(data[feature].mean()*1.05,max_ylim*0.96,'Média : {:.2f}'.format(data[feature].mean()))
```

# GRAFICO DE BARRAS

```python
for feature in data:
    plt.figure(figsize=(17,17))
    sns.set_theme(style='darkgrid')

    sns.countplot(x=feature, data=data)
```

# GRAFICO DE REGRESSÃO

```python

for feature in data:
    plt.figure(figsize=(10,10))
    sns.regplot(x=data['age'],y=feature,data=data)
```

```python
for feature in data:
    plt.figure(figsize=(15,15))
    sns.jointplot(x=data['age'],y=data[feature], data=data,kind='hex',color='green')
```
